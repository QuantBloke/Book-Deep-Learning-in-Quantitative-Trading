We here show examples of a limit order book (LOB). This is a simulated order book data that has the same structure of a real limit order book.
We include 10 levels on Ask and Bid sides of a LOB. 

```python
import os 
import sys
current_dir = os.path.dirname(os.path.abspath("__file__"))
project_dir = os.path.dirname(current_dir)
sys.path.append(project_dir)

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/limit_order_book_data.csv')
print(df.head())
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['AskPrice1'][:100], label='AskPrice1')
plt.plot(df['AskPrice2'][:100], label='AskPrice2')
plt.plot(df['AskPrice3'][:100], label='AskPrice3')

plt.plot(df['BidPrice1'][:100], label='BidPrice1')
plt.plot(df['BidPrice2'][:100], label='BidPrice2')
plt.plot(df['BidPrice3'][:100], label='BidPrice3')

plt.legend()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['AskVolume1'][:100], label='AskVolume1')
plt.plot(df['AskVolume2'][:100], label='AskVolume2')
plt.plot(df['AskVolume3'][:100], label='AskVolume3')

plt.plot(df['BidVolume1'][:100], label='BidVolume1')
plt.plot(df['BidVolume2'][:100], label='BidVolume2')
plt.plot(df['BidVolume3'][:100], label='BidVolume3')

plt.legend()
```
